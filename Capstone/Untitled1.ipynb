```python
from __future__ import division
from utils import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from functools import partial
%matplotlib inline
```

```python
def lret(x): 
    x = x - x.mean()
    return logreturn(x, x.shift(1))

def macross(x, lw,sw):
    ret = lret(x)
    l = ret.ewm(lw).mean()
    s = ret.ewm(sw).mean()
    return (s > l) * 1
```

```python
def load_data():   
    feature_defs = {}
    for s in [2,4,8,16,32,64]:
        l = 2*s
        feature_defs['ma_{}_{}'.format(l,s)] = lambda x:macross(x,l,s)    
    
    features_and_targets = partial(
        create_features_and_targets,
        feature_defs=feature_defs,
        target_defs={'up': lambda x:  (lret(x.shift(-1)) > 0)*1 }
    )
    return load_stock_datasets(features_and_targets)
```

```python
data = load_data()
ds = concatenate_datasets(data)
```
